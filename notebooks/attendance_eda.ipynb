```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

plt.figure(figsize=(20, 6))
sns.scatterplot(x='Opponent', y='Attendance', data=df_MF, s=200, palette='coolwarm')

# Function to add images as markers
def plot_images(x, y, image_path):
    for i in range(len(x)):
        img = plt.imread(image_path)
        imagebox = OffsetImage(img, zoom=0.05)  # Adjust the zoom factor as needed
        ab = AnnotationBbox(imagebox, (x[i], y[i]), frameon=False, pad=0)
        plt.gca().add_artist(ab)

# Get the x and y coordinates from the seaborn plot
temp = df_MF[df_MF['Opponent'] == 'NCCU']

x_coords = temp['Opponent']
y_coords = temp['Attendance']


base_path = 'logos_mf'

for opp in temp['Opponent'].unique():
    try:
        image_path_class = f"{os.getcwd()}\\{base_path}\\{opp}.png"
        plot_images(x_coords, y_coords, image_path_class)
    except:
        image_path_class = f"{os.getcwd()}\\{base_path}\\{opp}.jpg"
        plot_images(x_coords, y_coords, image_path_class)


# Call the function for each class
# plot_images(x_coords, y_coords, image_path_class1)
# Repeat for other classes if needed

# Set labels and title
plt.title('W/L V/S Attendance for Women\'s Basketball')
plt.xlabel('Opponent')
plt.ylabel('Attendance')

# Save and show the plot
plt.savefig('plt_images/wl_vs_attendance_wb.png')
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(x='Opponent', y='Attendance', data=df_MF, s=200, palette='coolwarm')

# Function to add images as markers
def plot_images(x, y, image_path):
    for i in range(len(x)):
        img = plt.imread(image_path)
        imagebox = OffsetImage(img, zoom=0.05)  # Adjust the zoom factor as needed
        ab = AnnotationBbox(imagebox, (x[i], y[i]), frameon=False, pad=0)
        plt.gca().add_artist(ab)

# Get the unique opponents in the DataFrame
opponents = df_MF['Opponent'].unique()

# Set the base path for logos
base_path = 'logos_mf'

# Iterate over each opponent
for opp in opponents[:2]:
    # Extract the data for the current opponent
    temp = df_MF[df_MF['Opponent'] == opp]
    
    # Get the x and y coordinates
    x_coords = temp['Opponent']
    y_coords = temp['Attendance']
    
    try:
        # Construct the full image path for the current opponent
        image_path_class = os.path.join(os.getcwd(), base_path, f"{opp}.png")
        # Call the function to add images as markers
        plot_images(x_coords, y_coords, image_path_class)
    except FileNotFoundError:
        # If the PNG file is not found, try with a JPG file
        image_path_class = os.path.join(os.getcwd(), base_path, f"{opp}.jpg")
        plot_images(x_coords, y_coords, image_path_class)

# Set labels and title
plt.title('W/L V/S Attendance for Women\'s Basketball')
plt.xlabel('Opponent')
plt.ylabel('Attendance')

# Save and show the plot
plt.savefig('plt_images/wl_vs_attendance_wb.png')
plt.show()
```
